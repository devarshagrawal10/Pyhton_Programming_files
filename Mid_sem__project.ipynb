```python
import datetime


buses = {}  # bus_id (key) -> {details}
routes = {}  # route_id (key) -> {details}
passengers = {}  # passenger_id (key) -> {details}

def add_bus():
    try:
        bus_id = int(input("Enter Bus ID: "))
        capacity = int(input("Enter Bus Capacity: "))
        driver_name = input("Enter Driver Name: ")
        buses[bus_id] = {'capacity': capacity, 'driver_name': driver_name}
        print("Bus added successfully!")
    except ValueError:
        print("Invalid input. Please enter integers for ID and capacity.")


def add_route():
    while True:
        try:
            route_id = int(input("Enter Route ID: "))
            origin = input("Enter Origin: ")
            destination = input("Enter Destination: ")
            stops = input("Enter Stops (comma-separated): ").split(",")
            break
        except ValueError:
            print("Invalid input. Please enter integers for ID.")
    routes[route_id] = {'origin': origin, 'destination': destination, 'stops': stops}
    print("Route added successfully!")


def add_passenger():
    while True:
        try:
            passenger_id = int(input("Enter Passenger ID: "))
            name = input("Enter Passenger Name: ")
            contact = input("Enter Passenger Contact: ")
            break
        except ValueError:
            print("Invalid input. Please enter integers for ID.")
        passengers[passenger_id] = {'name': name, 'contact': contact}
        print("Passenger added successfully!")


def book_ticket():
    while True:
        try:
            passenger_id = int(input("Enter Passenger ID: "))
            bus_id = int(input("Enter Bus ID: "))
            route_id = int(input("Enter Route ID: "))
            today = datetime.date.today().strftime("%Y-%m-%d")
            date = input(f"Enter Booking Date (format YYYY-MM-DD) [Today: {today}]: ") or today
            if not validate_date(date):
                print("Invalid date format. Please use YYYY-MM-DD.")
                continue
            if bus_id not in buses or route_id not in routes or passenger_id not in passengers:
                print("Invalid bus ID, route ID, or passenger ID. Please try again.")
                continue

            # Placeholder for booking logic (replace with your implementation)
            # (Example: Create a ticket or mark a seat as booked)
            print("Ticket booked successfully! (Booking logic needs implementation)")
            break
        except ValueError:
            print("Invalid input. Please enter integers for ID.")


def validate_date(date_text):
    try:
        datetime.datetime.strptime(date_text, "%Y-%m-%d")
        return True
    except ValueError:
        return False


def show_buses():
    if not buses:
        print("No buses found!")
        return
    for bus_id, details in buses.items():
        print(f"Bus ID: {bus_id}")
        print(f"Capacity: {details['capacity']}")
        print(f"Driver Name: {details['driver_name']}")
        print("-" * 20)


def show_routes():
    if not routes:
        print("No routes found!")
        return
    for route_id, details in routes.items():
        print(f"Route ID: {route_id}")
        print(f"Origin: {details['origin']}")
        print(f"Destination: {details['destination']}")
        print(f"Stops: {', '.join(details['stops'])}")
        print("-" * 20)


def show_passengers():
    if not passengers:
        print("No passengers found!")
        return
    for passenger_id, details in passengers.items():
        print(f"Passenger ID: {passenger_id}")
        print(f"Name: {details['name']}")
        print(f"Contact: {details['contact']}")
        print("-" * 20)  # Separator between passengers


def search_buses(origin, destination):

    found_buses = False

    for bus_id, details in buses.items():
        routes = details.get('routes', [])  # Get associated routes (if any)
        for route in routes:
            if route['origin'] == origin and route['destination'] == destination:
                print(f"Bus ID: {bus_id}")
                print(f"Driver Name: {details['driver_name']}")
                print("-" * 20)
                found_buses = True

def main():
    while True:
        print("\nBus Management System")
        print("1. Add Bus")
        print("2. Add Route")
        print("3. Add Passenger")
        print("4. Book Ticket")
        print("5. Show Buses")
        print("6. Show Routes")
        print("7. Show Passengers")
        print("8. Search Buses")
        print("9. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            add_bus()
        elif choice == '2':
            add_route()
        elif choice == '3':
            add_passenger()
        elif choice == '4':
            book_ticket()
        elif choice == '5':
            show_buses()
        elif choice == '6':
            show_routes()
        elif choice == '7':
            show_passengers()
        elif choice == '8':
            origin = input("Enter Origin: ")
            destination = input("Enter Destination: ")
            search_buses(origin, destination)
        elif choice == '9':
            print("Exiting Bus Management System!")
            break
        else:
            print("Invalid choice. Please try again.")


if _name_ == "_main_":
    main()
```
