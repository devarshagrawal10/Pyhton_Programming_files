```python
import datetime


buses = {}  # bus_id (key) -> {details}
routes = {}  # route_id (key) -> {details}
passengers = {}  # passenger_id (key) -> {details}


def add_bus():
    while True:
        try:
            bus_id = int(input("Enter Bus ID: "))
            capacity = int(input("Enter Bus Capacity: "))
            driver_name = input("Enter Driver Name: ")
            break
        except ValueError:
            print("Invalid input. Please enter integers for ID and capacity.")
    add_bus(bus_id, capacity, driver_name)
    print("Bus added successfully!")


def add_route():
    while True:
        try:
            route_id = int(input("Enter Route ID: "))
            origin = input("Enter Origin: ")
            destination = input("Enter Destination: ")
            stops = input("Enter Stops (comma-separated): ").split(",")
            break
        except ValueError:
            print("Invalid input. Please enter integers for ID.")
    add_route(route_id, origin, destination, stops)
    print("Route added successfully!")


def add_passenger():
    while True:
        try:
            passenger_id = int(input("Enter Passenger ID: "))
            name = input("Enter Passenger Name: ")
            contact = input("Enter Passenger Contact: ")
            break
        except ValueError:
            print("Invalid input. Please enter integers for ID.")
    add_passenger(passenger_id, name, contact)
    print("Passenger added successfully!")


def book_ticket():
    while True:
        try:
            passenger_id = int(input("Enter Passenger ID: "))
            bus_id = int(input("Enter Bus ID: "))
            route_id = int(input("Enter Route ID: "))
            today = datetime.date.today().strftime("%Y-%m-%d")
            date = input(f"Enter Booking Date (format YYYY-MM-DD) [Today: {today}]: ") or today
            if not validate_date(date):
                print("Invalid date format. Please use YYYY-MM-DD.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter integers for ID.")
    book_ticket(passenger_id, bus_id, route_id, date)


def validate_date(date_text):
    try:
        datetime.datetime.strptime(date_text, "%Y-%m-%d")
        return True
    except ValueError:
        return False


def show_buses():
    if not buses:
        print("No buses found!")
        return
    show_buses()


def show_routes():
    if not routes:
        print("No routes found!")
        return
    show_routes()


def show_passengers():
    if not passengers:
        print("No passengers found!")
        return
    show_passengers()


def search_buses():
    origin = input("Enter Origin: ")
    destination = input("Enter Destination: ")
    today = datetime.date.today().strftime("%Y-%m-%d")
    date = input(f"Enter Travel Date (format YYYY-MM-DD) [Today: {today}]: ") or today
    if not validate_date(date):
        print("Invalid date format. Please use YYYY-MM-DD.")
        return
    search_buses(origin, destination, date)


def main_menu():
    print("-" * 50)
    print("Bus Management System")
    print("-" * 50)
    print("1. Add Bus")
    print("2. Add Route")
    print("3. Add Passenger")
    print("4. Book Ticket")
    print("5. Show Buses")
    print("6. Show Routes")
    print("7. Show Passengers")
    print("8. Search Buses")
    print("9. Exit")
    print("-" * 50)

while True:
        try:
            choice = int(input("Enter your choice: "))
            break
        except ValueError:
            print("Invalid input. Please enter a number.")
            if choice == 1:
              add_bus()
            elif choice == 2:
                add_route()
            elif choice == 3:
                  add_passenger()
            elif choice ==4:
                    book_ticket()
            elif choice==5:
                      validate_date(date_text)
            elif choice==6:
                        show_buses()
            elif choice==7:
                          show_passengers()
            elif choice == 8:
                            search_buses()
            elif choice == 9:
              break
```
